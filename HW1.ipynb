# 練習問題

I. 次のような2期間モデルを考える。家計は以下の効用関数を、各期の予算制約を所与として最大化する

$$
  U(c_1,c_2) = u(c_1) + \beta E u(c_2)
$$

subject to

\begin{align*}
  c_{1} + a_{2} &= a_{1} \\
  c_{2} &= (1+r)a_{2} + w
\end{align*}

ここで、2期目の所得は、
$$
  w = \begin{cases} w_{H} \text{ with prob. } p \\
    w_{L} \text{ with prob. } 1-p 
  \end{cases}
$$
であるとする

1. 効用最大化のための一階の必要条件（オイラー方程式）を求めよ
2. 以下では、$a_{1}$を$a_{1}^{i}\in\{0.1,0.2,...,1.0\}$の10個のグリッドで近似する。また、$\beta=0.985^{30}, 1+r=1.025^{30}, w_{H} = 1.0, w_{L} = 0.0, p = 0.5, u(c)=\ln c$とする
  1. 状態変数$a_{1}$と操作変数$a_{2}$がともに離散の場合の数値解を求めよ（ヒント：$w_{H}=0$の場合と比べて、貯蓄関数はどのように変化するだろうか？）
  2. 操作変数$a_{2}$を連続とした場合の数値解を求めよ
3. 2.で$w_{L}=0.1$とした場合に、貯蓄関数はどのような形をとるだろうか（ヒント：$a_{1}=0.1$の場合に注意せよ）

II. I.の2期間モデルを無限期間に拡張してみよう。すなわち、家計は以下の効用関数を、各期の予算制約を所与として最大化する

$$
  U(c_1,c_2,\dots) = E_{1}\sum_{t=1}^{\infty}\beta^{t-1}u(c_t)
$$

subject to

$$
  c_{t} + a_{t+1} = (1+r)a_{t} + w_{t}
$$

for $t=1,2,...$.

ここで、各期$t=1,2,...$の所得は、
$$
  w_{t} = \begin{cases} w_{H} \text{ with prob. } p \\
    w_{L} \text{ with prob. } 1-p 
  \end{cases}
$$

であるとする

1. $(a_{t},w_{t})$を状態変数として、ベルマン方程式を求めよ
2. 効用最大化のための一階の必要条件（オイラー方程式）を求めよ
3. $a_{t}$を$a_{t}^{i}\in\{0.0,0.1,0.2,...,10.0\}$の101個のグリッドで近似する。また、$\beta=0.985, 1+r=1.01, w_{H} = 1.0, w_{L} = 0.1, p = 0.9, u(c)=\ln c$とする
  1. 状態変数$a_{t}$と操作変数$a_{t+1}$がともに離散の場合の数値解を求めよ
  2. 操作変数$a_{t+1}$を連続とした場合の数値解を求めよ
4. 以下のそれぞれのパラメータの変化が、貯蓄関数に与える影響を議論せよ
  1. $1+r$が上昇する場合
  2. $p$が低下する場合

III. 以下の準線形ニューケインジアン・モデルを考える

\begin{align*}
  \lambda_{t} &= \mathbb{E}_{t}\lambda_{t+1} \color{red}{+} \left(r_{n,t}-\mathbb{E}_{t}\pi_{t+1}-s_{t}\right) \\
  \lambda_{t} &= d_{c}c_{t} + d_{n}n_{t} \\
  \pi_{t} &= \kappa (c_{t} + \frac{N}{1-N}n_{t}) + \beta \mathbb{E}_{t}\pi_{t+1} \\
  y_{t} &= (1-g_{y})c_{t} + g_{y}g_{t} \\
  n_{t} &= y_{t} \\
  r_{n,t}^{*} &= r^{*} + \phi_{\pi} \mathbb{E}_{t}\pi_{t+1} \\
  r_{n,t} &= \max \{r_{n,t}^{*}, 0\}
\end{align*}

ここで、$d_{c} = \gamma(1-\sigma)-1, d_{n} = -(1-\gamma)(1-\sigma)N/(1-N)$とする

1. $s_{t}=r^{*}, g_{t}=g>0$とするとき、ゼロ金利制約はバインドせず、$r_{n,t} = r_{n,t}^{*}>0$が成り立つとする
  1. $y_{t}$を$g_{t}$の線形な関数$y_{t}=A_{y}g_{t}$として表せ（ヒント：未定係数法を使う）
  2. $\beta=0.99,\kappa=0.03,\phi_{\pi}=1.5,g_{y}=0.2,r^{*}=100\beta^{-1}-100,\sigma=2,\gamma=0.29,N=1/3$とするとき、財政乗数$dy_{t}/dg_{t}=A_{y}/g_{y}$を求めよ
  
2. 確率変数$(s_{t},g_{t})$はそれぞれ2つの値のいずれか$s_{t}\in\{s_{H},s_{L}\}, g_{t}\in\{g_{H},g_{L}\}$をとり、以下の遷移行列を持つ共通の確率過程に従うとする
$$
  P = \left[\begin{array}{cc}
  1-p_{H} & p_{H} \\ 
  1-p_{L} & p_{L} \end{array}\right]
$$
すなわち、状態$H$のとき$s_{t}=s_{H}=r^{*}, g_{t}=g_{H}$、状態$L$のとき$s_{t}=s_{L}, g_{t}=g_{L}$であるとする。また、状態$L$においてのみゼロ金利制約はバインドし、状態$H$においては財政支出は行わないものとする（$g_{H}=0$）
  1. $(\pi_{L},y_{L})$を$g_{L}$の線形な関数として表せ
  2. 1.B.で与えたパラメータに加えて、$s_{L}=-1.5,p_{H}=0.0,p_{L}=0.8$とするとき、財政乗数$dy_{L}/dg_{L}$を求めよ（ヒント：$g_{L}\in\{0.0,1.0\}$のそれぞれで$y$の値を求め、差分をとってみよう）
  3. $p_{L}$の値を変化させたとき、財政乗数はどのように変わるだろうか？

IV. II.で計算した政策関数$a'=g(a,w)$を用いて、$(a,w)$の定常分布を計算したい。今期の個人の状態変数を$(a_{k},w_{i})$とするとき、

  1. 操作変数$a'$が離散の場合、政策関数$a_{l}=g(a_{k},w_{i})$から、次期の状態変数が$(a_{l},w_{j})$にある確率は$p_{ij}$である。このとき、$(a,w)$の定常分布を計算せよ
  1. 操作変数$a'$が連続の場合、政策関数$a'=g(a_{k},w_{i})$から、$a' \in [a_{l},a_{l+1}]$であるとき、
    - 次期の状態変数が$(a_{l},w_{j})$である確率は$(1-\omega) \cdot p_{ij}$、
    - $(a_{l+1},w_{j})$である確率は$\omega \cdot p_{ij}$
    
  ここで、$\omega=\frac{a'-a_{l}}{a_{l+1}-a_{l}}$である。$(a,w)$の定常分布を計算せよ
  1. $w_{t}$が以下の遷移行列に従うマルコフ過程であるとき、定常分布はどのように変わるか。特に借入制約に直面する数の家計に着目して説明せよ
  $$
    P = \left[\begin{array}{cc} 0.9 & 0.1 \\
    0.5 & 0.5 \end{array}\right]
  $$

ヒント：2状態変数の推移確率行列$P$を所与として、$t$期の分布を$\pi_{t}$とすると、その遷移式は以下のように表される

$$
  \pi'_{t+1} = \pi'_{t} P
$$
